```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
x = np.array([5,12,20,30,40,50])
y = np.array([10,20,18,28,25,67])

X = x.reshape(-1,1)#coulumn vector
#imput must be in 2D
print(X)
```

### Model Creation

```python
model = LinearRegression()
model.fit(X,y)
```

```python
print("Slope: ", model.coef_)
print("Y-intercept: ",model.intercept_)
```

#### Displaying the line of best fit

```python
y_bar = model.coef_ * x + model.intercept_
```

```python
plt.scatter(x, y)
plt.plot(x,y_bar, color= 'red')
```

##### Scoring the model

```python
print(f"The accuracy of the model is: , {model.score(X, y): .2f}")
```

##### predict for unseen data

```python
x_new = np.array([95, 43, 82]).reshape(-1,1)
print("Prediction for new unseen data: ", model.predict(x_new))
```
