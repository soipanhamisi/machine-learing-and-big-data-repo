### linear regression on the make_wave dataset

```python
import mglearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
lr = LinearRegression().fit(X_train, y_train)
print("Lr.coef: {}".format(lr.coef_))
print("Lr.intercept: {}".format(lr.intercept_))
```

```python
print("Training set score: {:.2f}".format(lr.score(X_train, y_train)))
print("Testing set score: {:.2f}".format(lr.score(X_test, y_test)))
```

### linear regression on the more complex Boston dataset

```python
X, y = mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 0)
lr = LinearRegression().fit(X_train, y_train)
print("Training set score: {:.2f}".format(lr.score(X_train, y_train)))
print("Testing set score: {:.2f}".format(lr.score(X_test, y_test)))
```

```python
print("Lr.coef: {}".format(lr.coef_))
print("Lr.intercept: {}".format(lr.intercept_))
```

#### ridge regression on the Boston dataset
Ridge regression is similar to linear regression, but adds a penalty to shrink the model's coefficients toward zero, without ever setting them to exactly zero. For a dataset with only one feature, this means the slope of the regression line will be less steep (flatter) than the ordinary linear regression line, but not perfectly flat
Regularization → explicitly restricting a model to prevent overfitting

```python
from sklearn.linear_model import Ridge
rr = Ridge().fit(X_train, y_train)
print("Training Set score: {:.2f}".format(rr.score(X_train, y_train)))
print("Test Set score: {:.2f}".format(rr.score(X_test, y_test)))
```

### comparing scores between the training linear and ridge regression
Observation → rr model has lower training set accuracy than lr but higher test set accuracy than lr <br>
explanation → rr actively tries to prevent excess model complexity and overfitting by lowering the coefficient sizes.
